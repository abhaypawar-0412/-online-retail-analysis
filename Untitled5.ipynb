```python
# Ensure basic cleaning is done
df = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)]
df['TotalPrice'] = df['Quantity'] * df['UnitPrice']

# Create features per CustomerID
customer_df = df.groupby('CustomerID').agg({
    'InvoiceNo': 'nunique',     # Frequency
    'Quantity': 'sum',          # Total quantity
    'TotalPrice': 'sum'         # Total monetary
}).reset_index()

# Rename columns for clarity
customer_df.columns = ['CustomerID', 'Frequency', 'TotalQuantity', 'TotalSpend']

# Preview
customer_df.head()
```

```python
from sklearn.preprocessing import StandardScaler

# Select the features for clustering
features = ['Frequency', 'TotalQuantity', 'TotalSpend']
scaler = StandardScaler()
scaled_data = scaler.fit_transform(customer_df[features])
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertia = []
K_range = range(1, 11)

for k in K_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(scaled_data)
    inertia.append(kmeans.inertia_)

# Plot the elbow curve
plt.figure(figsize=(8,4))
plt.plot(K_range, inertia, 'bo-')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method For Optimal k')
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import KMeans

# Apply KMeans clustering
k = 4  # Number of clusters
kmeans = KMeans(n_clusters=k, random_state=42)
customer_df['Cluster'] = kmeans.fit_predict(scaled_data)

# Preview the result
customer_df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
sns.scatterplot(
    data=customer_df, 
    x='TotalSpend', 
    y='TotalQuantity', 
    hue='Cluster', 
    palette='tab10', 
    alpha=0.7,
    s=100
)

plt.title('Customer Segments by Spend & Quantity')
plt.xlabel('Total Spend')
plt.ylabel('Total Quantity Purchased')
plt.grid(True)
plt.legend(title='Cluster')
plt.tight_layout()
plt.show()
```

```python
# Cluster summary using groupby
cluster_summary = customer_df.groupby('Cluster').agg({
    'Frequency': 'mean',
    'TotalQuantity': 'mean',
    'TotalSpend': 'mean',
    'CustomerID': 'count'  # number of customers in each cluster
}).rename(columns={'CustomerID': 'NumCustomers'}).round(2)

cluster_summary
```

```python
import matplotlib.pyplot as plt

# Set figure size
plt.figure(figsize=(15,10))

# Total Spend
plt.subplot(2, 2, 1)
customer_df.groupby('Cluster')['TotalSpend'].mean().plot(kind='bar', color='green')
plt.title('Average Total Spend per Cluster')
plt.ylabel('Total Spend')

# Total Quantity
plt.subplot(2, 2, 2)
customer_df.groupby('Cluster')['TotalQuantity'].mean().plot(kind='bar', color='orange')
plt.title('Average Total Quantity per Cluster')
plt.ylabel('Total Quantity')

# Frequency
plt.subplot(2, 2, 3)
customer_df.groupby('Cluster')['Frequency'].mean().plot(kind='bar', color='skyblue')
plt.title('Average Frequency per Cluster')
plt.ylabel('Frequency')

# Number of Customers
plt.subplot(2, 2, 4)
customer_df['Cluster'].value_counts().sort_index().plot(kind='bar', color='purple')
plt.title('Number of Customers per Cluster')
plt.ylabel('Count')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Behavioral cluster labels
cluster_labels = {
    0: "High-Value Customers",
    1: "Frequent Low Spenders",
    2: "Infrequent Low Spenders",
    3: "Loyal Big Buyers"
}

# Map cluster numbers to behavior
customer_df['ClusterLabel'] = customer_df['Cluster'].map(cluster_labels)

# Plot with clear legend
plt.figure(figsize=(10,6))
scatter = sns.scatterplot(
    data=customer_df, 
    x='TotalSpend', 
    y='TotalQuantity', 
    hue='ClusterLabel', 
    palette='tab10', 
    alpha=0.7,
    s=100
)

plt.title('Customer Segments by Spend & Quantity\n(Legend: Customer Behavior)', fontsize=14)
plt.xlabel('Total Spend')
plt.ylabel('Total Quantity Purchased')
plt.grid(True)

# Custom legend title
plt.legend(title='Customer Behavior', title_fontsize='13', fontsize='11')

plt.tight_layout()
plt.savefig("ustomer_clusters_behavioral.png")
plt.show()
```
